```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
start = pd.read_csv("cmake-build-debug/start.csv", sep = ';', header=None)
finish = pd.read_csv("cmake-build-debug/finish.csv", sep = ';', header=None)
```

```python
plt.figure(figsize=(16, 16))
plt.scatter(start[0], start[1])
first = True
for i, x in start.iterrows():
    if first:
        prev_x = x
        first = False
    plt.plot([x[0], prev_x[0]], [x[1], prev_x[1]])
    prev_x = x
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(start[0], start[1])
first = True
for i, x in finish.iterrows():
    if first:
        prev_x = x
        first = False
    plt.plot([x[0], prev_x[0]], [x[1], prev_x[1]])
    prev_x = x
```
